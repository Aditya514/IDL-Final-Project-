```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
%cd /content/drive/MyDrive/IDL Project/Final-Submission/SSPNet-Speaker-Personality-Corpus/Personality_Scores
```

```python
df_labels=pd.read_csv('Score_001.csv')['Clip_ID']
df1=pd.read_csv('Score_001.csv').iloc[:,1:].to_numpy()
df2=pd.read_csv('Score_002.csv').iloc[:,1:].to_numpy()
df3=pd.read_csv('Score_003.csv').iloc[:,1:].to_numpy()
df4=pd.read_csv('Score_004.csv').iloc[:,1:].to_numpy()
df5=pd.read_csv('Score_005.csv').iloc[:,1:].to_numpy()
df6=pd.read_csv('Score_006.csv').iloc[:,1:].to_numpy()
df7=pd.read_csv('Score_007.csv').iloc[:,1:].to_numpy()
df8=pd.read_csv('Score_008.csv').iloc[:,1:].to_numpy()
df9=pd.read_csv('Score_009.csv').iloc[:,1:].to_numpy()
df10=pd.read_csv('Score_010.csv').iloc[:,1:].to_numpy()
df11=pd.read_csv('Score_011.csv').iloc[:,1:].to_numpy()
```

```python
df_mean=np.mean([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11],axis=0)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(df_mean)
```

```python
df_mean=scaler.transform(df_mean)
```

```python
df_all = pd.DataFrame(df_mean, columns=['Extraversion',	'Agreeableness',	'Conscientiousness',	'Neuroticism',	'Openness'])
df_all=pd.concat([df_labels,df_all],axis=1)
```
