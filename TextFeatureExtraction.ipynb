```python
import speech_recognition as sr
from sentence_transformers import SentenceTransformer
model =  SentenceTransformer("dangvantuan/sentence-camembert-large")
```

```python
wav_files=os.listdir("/content/drive/MyDrive/IDL Project/Final-Submission/SSPNet-Speaker-Personality-Corpus/Audio_clips")
```

```python
to_text = sr.Recognizer()
```

```python
text_embeddings=os.listdir("/content/drive/MyDrive/IDL Project/Final-Submission/Merged Dataset/Text Embeddings")
```

```python
len(text_embeddings)
```

```python
for i,wav_filename in enumerate(wav_files):
  filename=wav_filename.split(".")[0]
  print(i)
  if str(filename+'.npy') not in text_embeddings:
    try:
      ex1 = sr.AudioFile(str(wav_filename))
      with ex1 as source:
        audio = to_text.record(source)
      text=to_text.recognize_google(audio, language = 'fr-FR')
      text_embedding = model.encode(text)
    
      text_numpy=np.save("/content/drive/MyDrive/IDL Project/Final-Submission/Merged Dataset/Text Embeddings/{}.npy".format(filename),text_embedding)
    except:
      print("Error",filename)
```
